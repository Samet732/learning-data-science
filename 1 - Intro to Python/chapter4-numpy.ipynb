# Numpy

### Importing

```python
import numpy as np
```

### Create and Multipy Numpy Array

```python
l = [1, 2, 3]
l1 = [4, 5, 6]
np_list = np.array(l)
np_list1 = np.array(l1)
np_list2 = np_list * np_list1

np_list2
```

### Numpy Array Item Type

```python
# only contains same type
arr = np.array([1.5, True])
arr1 = np.array([1.5, True, "Samet"])
print(arr)
print(arr1)
```

### Array Dimensions

```python
two_d = np.array([[1,2,3], [1,2,3]])
print(two_d.shape)
print(two_d[:, 0:2])
```

### Generate Data

```python
height = np.random.normal(1.72, 0.20, 500)
weight = np.random.normal(65, 20, 500)

np_city = np.round(np.column_stack([height, weight]), 2)
np_city[:, 0]
```

### Comparing Numpy Array

```python
ages = np.array([11, 18, 19, 32, 7, 49, 23, 35])
print(ages > 18)
```

```python
print(ages[ages > 18])
```

```python
print(np.logical_and(ages > 18, ages < 30))
print(ages[np.logical_and(ages > 18, ages < 30)])
```

### Loop Arrays

```python
l = ["a", "b", "c"]
for index, val in enumerate(l):
    print(index, val)
```

#### Numpy Array

```python
for val in ages:
    print(val)
```

#### 2D Numpy Array

```python
for val in np_city:
    print(val)
```

```python
for val in np.nditer(np_city):
    print(val)
```

### Transpose

```python
# transpose numpy array
np_arr = np.array([[1, 3, 5], [2, 4, 6]])
print(np_arr)
np_arr_t = np.transpose(np_arr)
print(np_arr_t)
```

### Random

```python
np.random.seed(123)
print(np.random.rand())
print(np.random.randint(1, 5))
print(np.random.randint(1, 5))
# always first and second numbers equal 2 and 3, if you assign seed to 123
```
