# Matplotlib

### Importing

```python
import matplotlib.pyplot as plt 

year = [1950, 1951, 1952, 1975, 2100]  
pop = [2.538, 2.57, 2.62, 6.43, 10.85]
```

### Plots

```python
plt.plot(year, pop)
plt.show()
```

### Customizing

```python
plt.clf() # clears
plt.plot(year, pop)

plt.xlabel('Year')
plt.ylabel('Population')

plt.title("World Population Projections")
plt.yticks([0, 2, 4, 6, 8, 10], ['0B', '2B', '4B', '6B', '8B', '10B'])

plt.show()
```

### Scatters

```python
plt.clf()

plt.scatter(year, pop, s=pop, c='red', alpha=0.8)

plt.text(2030, 6, "Text")
plt.grid(True)

plt.show()
```

### Histograms

```python
plt.clf();
plt.hist(year, bins=3);
plt.show();
```
